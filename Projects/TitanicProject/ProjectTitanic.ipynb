```python
import pandas as pd
import re
```

```python
df = pd.read_csv('titanic.csv')
```

```python
filter1 = df.loc[(df['Sex'] == 'female') & (df['Survived'] == 1) & (df['Pclass'] == 1)] #Filtrando as informações corretas...
filter2 = filter1[['Name', 'Age', 'Sex', 'Pclass', 'Survived']] #Criando a tabela somente com o que ele quer...
filter2.sort_values('Name', ascending = True) #Ordem alfabética...
```

```python
#Criando o Arquivo:
filter2 = pd.DataFrame(filter2)
filter2.to_csv('titanic_results.csv', index=False)
```
