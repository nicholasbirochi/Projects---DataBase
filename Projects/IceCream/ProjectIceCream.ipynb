```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
icecream = pd.read_csv('IceCreamData.csv')
icecream = icecream.sort_values('Revenue', ascending = True)
```

```python
#Extra:
celsius = (icecream['Temperature']- 32)* 5/9
icecream['Temperature'] = celsius
```

```python
plt.title('Ice Cream Truck:')

plt.plot(icecream['Temperature'], icecream['Revenue'], 'r.--', label = 'US$')

plt.xlabel('Temperature (Fahrenheit)')
plt.ylabel('Sales (US$)')
plt.legend()

#Criando o Arquivo:
plt.savefig('icecream_sales1.pdf')
plt.show()
```
