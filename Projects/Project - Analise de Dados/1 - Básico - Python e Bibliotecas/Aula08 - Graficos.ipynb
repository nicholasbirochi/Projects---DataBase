```python
import matplotlib.pyplot as plt
```

```python
nomes = ['Rafael', 'Maria', 'Ana']
idades = [10, 20, 15]

#Gráfico de linhas:
plt.plot(nomes,idades)
```

```python
#Gráfico de barras, é mais usual criar uma imagem na hora de criar os gráficos, já que assim você tem mais controle sobre o seu gráfico, maneira certa:
plt.figure(figsize = (5, 5))
plt.bar(nomes,idades)
plt.xlabel('Nome')
plt.ylabel('Idade')
```
